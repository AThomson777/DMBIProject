### Training A Model

```python
#Next it is time to label encode both Electrode and the Target. Since there are 32 electrodes, we will label encode for simplicity sake
# Though if I had more computational power I would perhaps do dummies

final_dataset['Electrode_Code'] = pd.factorize(final_dataset['Electrode'])[0]

final_dataset
```

```python
final_dataset['Target_Code'] = pd.factorize(final_dataset['Target'])[0]

final_dataset
```

```python
final_dataset = final_dataset.drop(columns=['Electrode','Target'])

final_dataset
```

```python
# While this dataset is hard to read it is much more machine readable for the computer

X = final_dataset.drop(columns=['Target_Code'])
y = final_dataset['Target_Code']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=77)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
```
